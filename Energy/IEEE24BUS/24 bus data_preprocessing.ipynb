# Visualization of IEEE 24 Bus System, original data

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

def parse_raw_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    bus_data = []
    load_data = []
    generator_data = []
    branch_data = []
    
    section = None
    for line in lines:
        line = line.strip()
        if "0 / END OF BUS DATA" in line:
            section = None
            continue
        elif "0 / END OF LOAD DATA" in line:
            section = None
            continue
        elif "0 / END OF GENERATOR DATA" in line:
            section = None
            continue
        elif "0 / END OF BRANCH DATA" in line:
            section = None
            continue
        elif "BEGIN LOAD DATA" in line:
            section = "LOAD"
            continue
        elif "BEGIN GENERATOR DATA" in line:
            section = "GENERATOR"
            continue
        elif "BEGIN BRANCH DATA" in line:
            section = "BRANCH"
            continue
        elif "0,    100.00" in line:  # Indicator of the start of bus data
            section = "BUS"
            continue
        
        if section == "BUS" and line:
            fields = line.split(',')
            if len(fields) >= 9:  # Ensure there are enough fields to parse
                bus_data.append({
                    "Bus ID": int(fields[0].strip()),
                    "Voltage (kV)": float(fields[2].strip()),
                    "Active Load (MW)": float(fields[7].replace("'", "").strip()),
                    "Reactive Load (MVAr)": float(fields[8].replace("'", "").strip())
                })
        elif section == "LOAD" and line:
            fields = line.split(',')
            if len(fields) >= 7:  # Ensure there are enough fields to parse
                load_data.append({
                    "Bus ID": int(fields[0].strip()),
                    "Active Load (MW)": float(fields[5].replace("'", "").strip()),
                    "Reactive Load (MVAr)": float(fields[6].replace("'", "").strip())
                })
        elif section == "GENERATOR" and line:
            fields = line.split(',')
            if len(fields) >= 2:  # Ensure there are enough fields to parse
                generator_data.append({
                    "Bus ID": int(fields[0].strip()),
                    "Active Power (MW)": float(fields[1].replace("'", "").strip())
                })
        elif section == "BRANCH" and line:
            fields = line.split(',')
            if len(fields) >= 2:  # Ensure there are enough fields to parse
                branch_data.append({
                    "From Bus": int(fields[0].strip()),
                    "To Bus": int(fields[1].strip())
                })
    
    return pd.DataFrame(bus_data), pd.DataFrame(load_data), pd.DataFrame(generator_data), pd.DataFrame(branch_data)

# Usage
bus_data, load_data, generator_data, branch_data = parse_raw_file("/Users/chris/Downloads/IEEE 24 bus DATA.RAW")
```
